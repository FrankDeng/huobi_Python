```python
def load_market(symbol,interval):
    market_client=MarketClient(init_log=True)
    list_obj=market_client.get_candlestick(symbol,interval,2000)
    data=pd.DataFrame(columns=['time','high','low','open','close','amount','volume'])    
    for item in list_obj:
        temp={'time':item.id,'high':item.high,'low':item.low,'open':item.open,'close':item.close,'amount':item.amount,'volume':item.vol}
        data=data.append(temp,ignore_index=True)
    return data

def load_trade(symbol):
    market_client = MarketClient()
    list_obj = market_client.get_history_trade("btcusdt", 1)
    data=pd.DataFrame(columns=['id','time','price','amount','direction'])  
    for item in list_obj:
        temp={'id':item.trade_id,'time':item.ts,'price':item.price,'amount':item.amount,'direction':item.direction}
        data=data.append(temp,ignore_index=True)
    return data

def update_market(data,symbol,interval):
    conn = sqlite3.connect('marketdata.db',timeout=10)
    ticker=symbol+"_"+interval
    c = conn.cursor()
    #get the count of tables with the name
    c.execute("SELECT count(name) FROM sqlite_master WHERE type='table' AND name='"+ticker+"'")
    # #if the count is 1, then table exists
    if c.fetchone()[0]==1:
        last_data=pd.read_sql('select * from '+ticker,conn)
        last_data=last_data.append(data).drop_duplicates().sort_values(by='time',ascending=True)
        last_data.sort_values(by='time',ascending=True).to_sql(ticker,conn,if_exists='replace', index=False)
    else:
        data.sort_values(by='time',ascending=True).to_sql(ticker,conn,if_exists='replace', index=False)
        
def read_market(symbol,interval):
    conn=sqlite3.connect('marketdata.db',timeout=10)
    ticker=symbol+"_"+interval
    data=pd.read_sql('select * from '+ticker,conn)
    return data
```

```python
def BackTest(data,method,display=False):
    eps = np.finfo(float).eps
    data['return']=data['close'].pct_change(1)
    data['factorChange']=data['factor'].diff()
    data['factorNorm']=(data['factor']-data['factor'].rolling(20).mean()).round(4)/(eps+data['factor'].rolling(20).std())
    data['factorMA']=data['factor'].rolling(5).mean()
    if method=="sigmoid":
        data['signal_Factor']=((data['factor']>0).astype(int)-(data['factor']<0).astype(int))
        data['signal_Change']=((data['factorChange']>0).astype(int)-(data['factorChange']<0).astype(int))
        data['signal_Norm']=((data['factorNorm']>0).astype(int)-(data['factorNorm']<0).astype(int))
        data['signal_MA']=(data['factorMA']>0).astype(int)-(data['factorMA']<0).astype(int)
        for x in ['Factor','Change','Norm','MA']:
            data['return_'+x]=(data['signal_'+x].replace(to_replace=0, method='ffill').shift(1)*data['return']).fillna(0).cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['position_'+x]=data['signal_'+x].cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['totalreturn_'+x]=(data['position_'+x].shift(1)*data['return']).cumsum()
    if method=="momentum":
        data['signal_Factor']=((data['factor'].diff().fillna(0)>0).astype(int)-(data['factor'].diff().fillna(0)<0).astype(int))
        data['signal_Change']=((data['factorChange'].diff().fillna(0)>0).astype(int)-(data['factorChange'].diff().fillna(0)<0).astype(int))
        data['signal_Norm']=((data['factorNorm'].diff().fillna(0)>0).astype(int)-(data['factorNorm'].diff().fillna(0)<0).astype(int))
        data['signal_MA']=((data['factorMA'].diff().fillna(0)>0).astype(int)-(data['factorMA'].diff().fillna(0)<0).astype(int))
        for x in ['Factor','Change','Norm','MA']:
            data['return_'+x]=(data['signal_'+x].replace(to_replace=0, method='ffill').shift(1)*data['return']).fillna(0).cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['position_'+x]=data['signal_'+x].cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['totalreturn_'+x]=(data['position_'+x].shift(1)*data['return']).cumsum()
    if method=="reversal":
        data['signal_Factor']=(data['factor']>data['factor'].shift(1))*(data['factor'].shift(1)<data['factor'].shift(2)).astype(int)-(data['factor']<data['factor'].shift(1))*(data['factor'].shift(1)>data['factor'].shift(2)).astype(int)
        data['signal_Change']=(data['factorChange']>data['factorChange'].shift(1))*(data['factorChange'].shift(1)<data['factorChange'].shift(2)).astype(int)-(data['factorChange']<data['factorChange'].shift(1))*(data['factorChange'].shift(1)>data['factorChange'].shift(2)).astype(int)
        data['signal_Norm']=(data['factorNorm']>data['factorNorm'].shift(1))*(data['factorNorm'].shift(1)<data['factorNorm'].shift(2)).astype(int)-(data['factorNorm']<data['factorNorm'].shift(1))*(data['factorNorm'].shift(1)>data['factorNorm'].shift(2)).astype(int)
        data['signal_MA']=(data['factorMA']>data['factorMA'].shift(1))*(data['factorMA'].shift(1)<data['factorMA'].shift(2)).astype(int)-(data['factorMA']<data['factorMA'].shift(1))*(data['factorMA'].shift(1)>data['factorMA'].shift(2)).astype(int)
        for x in ['Factor','Change','Norm','MA']:
            data['return_'+x]=(data['signal_'+x].replace(to_replace=0, method='ffill').shift(1)*data['return']).fillna(0).cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['position_'+x]=data['signal_'+x].cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['totalreturn_'+x]=(data['position_'+x].shift(1)*data['return']).cumsum()
    if method=="breakout_1":
        data['signal_Factor']=(data['factor']>(data['factor'].rolling(20).mean()+data['factor'].rolling(20).std())).astype(int)-(data['factor']<(data['factor'].rolling(20).mean()-data['factor'].rolling(20).std())).astype(int)
        data['signal_Change']=(data['factorChange']>(data['factorChange'].rolling(20).mean()+data['factorChange'].rolling(20).std())).astype(int)-(data['factorChange']<(data['factorChange'].rolling(20).mean()-data['factorChange'].rolling(20).std())).astype(int)
        data['signal_Norm']=(data['factorNorm']>(data['factorNorm'].rolling(20).mean()+data['factorNorm'].rolling(20).std())).astype(int)-(data['factorNorm']<(data['factorNorm'].rolling(20).mean()-data['factorNorm'].rolling(20).std())).astype(int)
        data['signal_MA']=(data['factorMA']>(data['factorMA'].rolling(20).mean()+data['factorMA'].rolling(20).std())).astype(int)-(data['factorMA']<(data['factorMA'].rolling(20).mean()-data['factorMA'].rolling(20).std())).astype(int)
        for x in ['Factor','Change','Norm','MA']:
            data['return_'+x]=(data['signal_'+x].replace(to_replace=0, method='ffill').shift(1)*data['return']).fillna(0).cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['position_'+x]=data['signal_'+x].cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['totalreturn_'+x]=(data['position_'+x].shift(1)*data['return']).cumsum()             
    if method=="breakout_2":
        data['signal_Factor']=(data['factor']>(data['factor'].rolling(20).mean()+data['factor'].rolling(20).std()*2)).astype(int)-(data['factor']<(data['factor'].rolling(20).mean()-data['factor'].rolling(20).std()*2)).astype(int)
        data['signal_Change']=(data['factorChange']>(data['factorChange'].rolling(20).mean()+data['factorChange'].rolling(20).std()*2)).astype(int)-(data['factorChange']<(data['factorChange'].rolling(20).mean()-data['factorChange'].rolling(20).std()*2)).astype(int)
        data['signal_Norm']=(data['factorNorm']>(data['factorNorm'].rolling(20).mean()+data['factorNorm'].rolling(20).std()*2)).astype(int)-(data['factorNorm']<(data['factorNorm'].rolling(20).mean()-data['factorNorm'].rolling(20).std()*2)).astype(int)
        data['signal_MA']=(data['factorMA']>(data['factorMA'].rolling(20).mean()+data['factorMA'].rolling(20).std()*2)).astype(int)-(data['factorMA']<(data['factorMA'].rolling(20).mean()-data['factorMA'].rolling(20).std()*2)).astype(int)
        for x in ['Factor','Change','Norm','MA']:
            data['return_'+x]=(data['signal_'+x].replace(to_replace=0, method='ffill').shift(1)*data['return']).fillna(0).cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['position_'+x]=data['signal_'+x].cumsum()
        for x in ['Factor','Change','Norm','MA']:
            data['totalreturn_'+x]=(data['position_'+x].shift(1)*data['return']).cumsum()
    result=data.copy()
    if display==True:
        Chart(result)
        Analytics(result,"Factor")
        Analytics(result,"MA")
        Analytics(result,"Norm")
        Analytics(result,"Change")
    return result

def Analytics(data,method,display=True):
    position=data["signal_"+method].replace(to_replace=0, method='ffill')
    period=len(data)
    turnover=position.diff().astype(bool).sum(axis=0)-1
    tradePnL=(position.diff()>0)*1.0-1.0*(position.diff()<0)
    tradePnL.iloc[0]=position.iloc[0]
    tradePnL=(tradePnL*data['close']).replace(to_replace=0,method='ffill')*(position.shift(1).astype(bool))
    tradePnL=-(tradePnL.shift(1)+tradePnL)*(position.diff()!=0)/data['close']
    win=(tradePnL>0).sum()
    averageWin=((tradePnL>0)*tradePnL).sum()/win
    lose=(tradePnL<0).sum()
    averageLose=((tradePnL<0)*tradePnL).sum()/lose
    totalRt=data['return_'+method].to_list()[-1]
    sharp=totalRt/data['return_'+method].diff().std()/math.sqrt(period)
    currentPosition=position.to_list()[-1]
    lastPosition=position.to_list()[-2]
    if display:
        print("Return_"+method)
        print("Period："+str(period))
        print("Turnover:"+str(turnover))
        print("Wins:"+str(win))
        print("AverageWin:{0:.2f}%".format(averageWin*100))
        print("Loses:"+str(lose))
        print("AverageLose:{0:.2f}%".format(averageLose*100))
        print("Total Return:{0:.2f}%".format(totalRt*100))
        print("Sharp:{0:.2f}%".format(sharp*100))
        print("currentPosition:"+str(currentPosition))
    if method=="Factor":
        lastfactors=data['factor'].iloc[-10:]
    else:
        lastfactors=data['factor'+method].iloc[-10:]
    stats=pd.DataFrame([period,turnover,win,averageWin,lose,averageLose,totalRt,sharp,lastPosition,currentPosition],
                 ['Period','Turnover','Wins','AvgWin','Loses','AvgLose','TotalRt','Sharp','lastPosition','currentPosition'])
    return stats

def Chart(data):
    width=.8
    fig, ax1 = plt.subplots()
    fig.set_size_inches(10,7.5)
    fig.set_dpi(80)
    lns1=ax1.plot(data['time'],data['return_Factor'],label="Factor",color="#ffa600")
    lns2=ax1.plot(data['time'],data['return_MA'],label="MA",color="#ff6361")
    lns3=ax1.plot(data['time'],data['return_Change'],label="Change",color="#bc5090")
    lns4=ax1.plot(data['time'],data['return_Norm'],label="Norm",color="#58508d")
    ax2 = ax1.twinx()
    lns5=ax2.plot(data['time'],data['close'],linestyle="--",color="#7d9fad",label="Price")
    lns = lns1+lns2+lns3+lns4+lns5
    labs = [l.get_label() for l in lns]
    ax1.legend(lns, labs, loc="upper left")
```
